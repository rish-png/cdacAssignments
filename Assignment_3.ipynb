#### Q2: Write a Pandas program to rename columns of a given DataFrame.

```Sample data:
Original DataFrame
   col1 col2 col3
0 1 4 7
1 2 5 8
2 3 6 9
New DataFrame after renaming columns:
  Column1 Column2 Column3
0 1 4 7
1 2 5 8
2 3 6 9```

````python
data_dict = { "col1": [0, 1, 4, 7],
              "col2": [1, 2, 5, 8],
              "col3": [2, 3, 6, 8]
            }

df1 = pd.DataFrame(data_dict)
print("Before Renaming Columns")
display(df1)
df1.rename(columns={"col1": "Column1", "col2": "Column2", "col3": "Column3"}, inplace=True)
print("\nAfter Renaming Columns")
display(df1)
````

#### Q5: Write a Pandas program to count city wise number of people from a given of data set (city, name of the person).

````python
person_lst = ["Sundar Pichai", "Larry Ellison", "Geoffrey Hinton", "Andrew Ng", "Ian Goodfellow", "Sheryl Sandberg", "Jack Dorsey", "Ilya Sutskever"]
residence = ["Venkatanarasimharajuvaripeta, Tamil Nadu", "Bhattu, Haryana", "Bhabua, Bihar", "Lailapur, Assam", "Bhattu, Haryana", "Bhattu, Haryana", "Bhabua, Bihar", "Bhabua, Bihar"]
city_people_dict = { "person": person_lst,
                     "city": residence
}
df2 = pd.DataFrame(city_people_dict)
print("DataFrame in use:")
display(df2)
print("\nCity-wise Count:")
display(df2.groupby('city').count())
````

#### Q6: Write a Pandas program to replace all the NaN values with Zero's in a column of a dataframe.

````python
data = { "someColumn1" : [np.nan, 1, 5.8, 9.5],
         "someColumn2" : [np.nan, 8, np.nan, 9.5],
         "someColumn3" : [99, 8, 55.56, 9.5],
         "someColumn4" : [7, np.inf, 5855.56, np.nan]
}
df3 = pd.DataFrame(data)
print("Before NAN update")
display(df3)
print("After NAN update")
display(df3.fillna(0))
````
