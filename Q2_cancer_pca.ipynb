### Use Breast Cancer Dataset and perform the following:
1. Implement Pre-processing functions
    - check for NULL values in the dataset. If any NULL value is present then treat it accordingly.
    - extract only feature columns from the dataset.
2. Apply Principal component analysis and extract 2 principal components from the dataset. 
3. Visualize the PCA components of the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer

dataset = load_breast_cancer()

X = dataset['data']
y = dataset['target']
```

```python
# print(dataset['DESCR'])
```

```python
print(X.shape, "\t", y.shape)
```

```python
from sklearn.preprocessing import StandardScaler

X = StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3, random_state=1)
```

```python
pca.fit(X)
```

```python
pca.explained_variance_ratio_.sum()
```

```python
X_pcs = pca.transform(X)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_pcs[:, 0], X_pcs[:, 1], X_pcs[:, 2], s=50, alpha=0.5, c=y, cmap='rainbow')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('Breast Cancer')
plt.show()
```
