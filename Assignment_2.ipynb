#### Q1: Write a Pandas program to create and display a DataFrame from a specified dictionary data which has the index labels.
```Sample DataFrame:```
```
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

````python
import pandas as pd
import numpy as np

exam_data = {'name': ['Alexander', 'Basil', 'Caesar', 'Demetri', 'Ellis', 'Fedora', 'Gabriella', 'Hercules', 'Igorr', 'Julius'],
             'score': [np.inf, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
             'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
            }
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df1 = pd.DataFrame(data=exam_data, index=labels)
display(df1)
````

#### Q13: Write a Pandas program to append a new row 'k' to DataFrame with given values for each column. Now delete the new row and return the original data frame.

````python
display("Before Addition of row:", df1)
df1.loc['k'] = ['Kyros', 19.5, 4, 'yes']
display("After Addition of row:", df1)
df1.drop('k', inplace=True)
display("After Deletion of row:", df1)
````

#### Q14: Write a Pandas program to sort the data frame first by 'name' in descending order, then by 'score' in ascending order.

````python
df1.sort_values(by=['name', 'score'], ascending=[False, True], inplace=True)
display(df1)
````

#### Q15: Write a Pandas program to replace the 'qualify' column contains the values 'yes' and 'no' with True and False.

````python
df1.replace(to_replace=['yes', 'no'], value=[True, False], inplace=True)
display(df1)
````

#### Q16: Write a Pandas program to change the name 'Julius' to 'Juno' in name column of the data frame.
