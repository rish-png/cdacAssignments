### Using the dataset of Q7. Impute some noise in the dataset. Visualize the dataset with and without noise. Then apply PCA on noisy dataset to remove the noise.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits

dataset = load_digits()

X = dataset['data']
y = dataset['target']
```

```python
# Adding noise to images

noise = np.random.randint(0, 15, X.shape)
X_mod = X + noise
```

```python
true_image = X[0].reshape(8,8)
noise_filled_image = X_mod[0].reshape(8,8)

figure = plt.figure(figsize=(4,2))
plt.subplot(1, 2, 1)
plt.imshow(true_image, cmap='binary')
plt.subplot(1,2, 2)
plt.imshow(noise_filled_image, cmap='binary')
plt.title("Original Image v/s Noise-filled Image", loc='right')
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=16, random_state=1)

X_pcs = pca.fit_transform(X_mod)
X_inversed = pca.inverse_transform(X_pcs)
```

```python
figure = plt.figure(figsize=(20,6))

for i in range(0,10):
    true_image = X[i].reshape(8,8)
    noise_image = X_mod[i].reshape(8,8)
    pca_image = X_inversed[i].reshape(8,8)

    plt.subplot(3, 10, i+1)
    plt.imshow(true_image, cmap='binary')
    
    plt.subplot(3, 10, i+11)
    plt.imshow(noise_image, cmap='binary')
    
    plt.subplot(3, 10, i+21)
    plt.imshow(pca_image, cmap='binary')

plt.show()

# 0th row - Actual Images
# 1st row - Noise injected Images
# 0th row - Noise reduced images (using PCA)
```
